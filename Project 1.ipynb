```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in csv
birth_path = 'Database 1.xlsx - Sheet2_starter.csv'
birth_data_df = pd.read_csv(birth_path)

birth_data_df
```

```python
birth_data_2012_df = birth_data_df.loc[birth_data_df['Year'] == 2015]
birth_data_2012_df
```

```python
birth_data_syd_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Sydney"]
birth_data_syd_df
```

```python
birth_data_nsw_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of NSW"]
birth_data_nsw_df
```

```python
birth_data_aus_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "AUSTRALIA"]
birth_data_aus_df
```

```python
# Greater Hobart
birth_data_hobart_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Hobart"]
# birth_data_hobart_df
```

```python
# Regional Tasmania
birth_data_tas_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of Tas."]
# birth_data_tas_df
```

```python
# Plot median age of mother Tasmainia: Regional vs Metro
plt.plot(birth_data_hobart_df["Year"], birth_data_hobart_df["Median age of mother"], label = "Greater Hobart")
plt.plot(birth_data_tas_df["Year"], birth_data_tas_df["Median age of mother"], color = 'r', label = "Regional Tasmania")
plt.title("Tasmania & Greater Hobart: Median age of mother 2012-2019")
plt.xlabel("Year")
plt.ylabel("Mother's median age")
plt.legend()
plt.show()
```

```python
# Greater Darwin
birth_data_darwin_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Darwin"]
# birth_data_darwin_df
```

```python
# Regional Northern Terriory
birth_data_nt_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of NT"]
# birth_data_nt_df
```

```python
# Plot median age of mother
plt.plot(birth_data_darwin_df["Year"], birth_data_darwin_df["Median age of mother"], color = 'g', label = "Greater Darwin")
plt.plot(birth_data_nt_df["Year"], birth_data_nt_df["Median age of mother"], color = 'm', label = "Regional NT")
plt.title("Greater Darwin: Median age of mother 2012-2019")
plt.xlabel("Year")
plt.ylabel("Mother's median age")
plt.legend()
plt.show()
```
