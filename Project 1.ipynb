```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in csv
birth_path = 'Database 1.xlsx - Sheet2_starter.csv'
birth_data_df = pd.read_csv(birth_path)

birth_data_df
```

```python
birth_data_2012_df = birth_data_df.loc[birth_data_df['Year'] == 2015]
birth_data_2012_df
```

```python
birth_data_syd_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Sydney"]
birth_data_syd_df
```

```python
birth_data_nsw_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of NSW"]
birth_data_nsw_df
```

```python
birth_data_aus_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "AUSTRALIA"]
birth_data_aus_df
```

```python
aus_birth_stats_df = birth_data_aus_df.describe()
aus_birth_stats_df
```

```python
plt.bar(aus_birth_stats_df[""])
```

```python
# Plot change in fertility rate of Australian mothers of all age categories from 2012 to 2019
plt.figure(figsize=(10,10))

plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["15–19 years"], color = 'orange', label = "15–19 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["20-24 years"], color = 'r', label = "20-24 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["25-29 years"], color = 'green', label = "25-29 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["30-34 years"], color = 'blueviolet', label = "30-34 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["35-39 years"], color = 'lightseagreen', label = "35-39 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["40-44 years"], color = 'navy', label = "40-44 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["45-49 years"], color = 'deeppink', label = "45-49 years")


plt.title("Changes over time in fertility rate of Australian mothers of all ages")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.legend(title = 'Age groups', bbox_to_anchor=(1.05, 1.0), loc = 'best')
plt.show()
```

```python
# Greater Hobart
birth_data_hobart_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Hobart"]
# birth_data_hobart_df
```

```python
# Regional Tasmania
birth_data_tas_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of Tas."]
# birth_data_tas_df
```

```python
# Plot median age of mother Tasmainia: Regional vs Metro
plt.plot(birth_data_hobart_df["Year"], birth_data_hobart_df["Median age of mother"], color = 'b', label = "Greater Hobart")
plt.plot(birth_data_tas_df["Year"], birth_data_tas_df["Median age of mother"], color = 'r', label = "Regional Tasmania")
plt.title("Tasmania & Greater Hobart: Median age of mother 2012-2019")
plt.xlabel("Year")
plt.ylabel("Mother's median age")
plt.legend()
plt.show()
```
