```python
# Linear regression model (line equation) for median age of mother and year

darwin_xaxis = birth_data_darwin_df.iloc[:,0]
darwin_yaxis = birth_data_darwin_df.iloc[:,12]
darwin_r_value = st.pearsonr(darwin_xaxis, darwin_yaxis)

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(darwin_xaxis, darwin_yaxis)

darwin_line_eq = "y= " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(darwin_line_eq)

darwin_regress_values = (slope * darwin_xaxis) + intercept

plt.plot(birth_data_darwin_df["Year"], birth_data_darwin_df["Median age of mother"], color = 'g', label = "Greater Darwin")
# plt.plot(birth_data_nt_df["Year"], birth_data_nt_df["Median age of mother"], color = 'm', label = "Regional NT")
plt.plot(darwin_xaxis, darwin_regress_values,"r-")
plt.title("Greater Darwin: Median age of mother 2012-2019")
plt.xlabel("Year")
plt.ylabel("Mother's median age")
plt.legend()
plt.show()
```
