```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
# Read in csv
birth_path = 'Database 1.xlsx - Sheet2_starter.csv'
birth_data_df = pd.read_csv(birth_path)

birth_data_df
```

```python
birth_data_adl_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Adelaide"]
birth_data_adl_df
```

```python
birth_data_sa_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of SA"]
birth_data_sa_df
```

```python
birth_data_melb_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Melbourne"]
birth_data_melb_df
```

```python
birth_data_vic_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of Vic."]
birth_data_vic_df
```

```python
# Plot median age of mother Victoria: Regional vs Metro
plt.plot(birth_data_melb_df["Year"], birth_data_melb_df["Median age of mother"], color = 'b', label = "Greater Melbourne")
plt.plot(birth_data_vic_df["Year"], birth_data_vic_df["Median age of mother"], color = 'r', label = "Regional Victoria")
plt.title("Victoria & Greater Melbourne: Median age of mother 2012-2019")
plt.xlabel("Year")
plt.ylabel("Mother's median age")
plt.legend()
plt.show()
```

```python
# Plot median age of mother South Australia: Regional vs Metro
plt.plot(birth_data_adl_df["Year"], birth_data_adl_df["Median age of mother"], color = 'b', label = "Greater Adelaide")
plt.plot(birth_data_sa_df["Year"], birth_data_sa_df["Median age of mother"], color = 'r', label = "Regional South Australia")
plt.title("South Australia & Greater Adelaide: Median age of mother 2012-2019")
plt.xlabel("Year")
plt.ylabel("Mother's median age")
plt.legend()
plt.show()
```

```python
birth_data_2019_df = birth_data_df.loc[birth_data_df['Year'] == 2019]
birth_data_2019_df
```

```python
birth_pop_2019_df = birth_data_2019_df.loc[:,["Place of Usual Residence","Births", "Estimated resident population"]]
#birth_pop_2019_df["Crude Birth Rate"]=""

birth_pop_2019_df
```

```python
#birth_pop_2019_df["Births"] = birth_pop_2019_df["Births"].astype(str).astype(int)
#birth_pop_2019_df["Estimated resident population"] = pd.to_numeric(birth_pop_2019_df["Estimated resident population"])
birth_pop_2019_df['Births'] = birth_pop_2019_df['Births'].str.replace(',', '')
birth_pop_2019_df['Estimated resident population'] = birth_pop_2019_df['Estimated resident population'].str.replace(',', '')
#birth_pop_2019_df.dtypes
#births_integer = birth_pop_2019_df["Births"].str.replace(',', '').astype(float).astype(int)
```
