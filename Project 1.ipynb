```python
# Plot change in fertility rate of Australian mothers of all age categories from 2012 to 2019
plt.figure(figsize=(6,5))

plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["15–19 years"], color = 'orange', label = "15–19 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["20-24 years"], color = 'r', label = "20-24 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["25-29 years"], color = 'green', label = "25-29 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["30-34 years"], color = 'blueviolet', label = "30-34 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["35-39 years"], color = 'lightseagreen', label = "35-39 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["40-44 years"], color = 'navy', label = "40-44 years")
plt.plot(birth_data_aus_df["Year"], birth_data_aus_df["45-49 years"], color = 'deeppink', label = "45-49 years")


plt.title("Changes over time in fertility rate of Australian mothers of all ages")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.legend(title = 'Age groups', bbox_to_anchor=(1.05, 1.0), loc = 'best')
plt.show()
```
