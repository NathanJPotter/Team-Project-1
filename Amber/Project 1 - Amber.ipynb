```python
#CALCULATE CRUDE BIRTH RATE
#Limit data to 2019
birth_data_2019_df = birth_data_df.loc[birth_data_df['Year'] == 2019]
#birth_data_2019_df

#Limit data to 2012
birth_data_2012_df = birth_data_df.loc[birth_data_df['Year'] == 2012]
#birth_data_2012_df
```

```python
#Focus on the beginning and end year of the time period

# Get columns needed for analysis, i.e. 'Place of Usual Residence', 'Births','Estimated Resident Population'
birth_pop_2019_df = birth_data_2019_df.loc[:,["Place of Usual Residence","Births", "Estimated resident population"]]
#birth_pop_2019_df

# Get columns needed for analysis, i.e. 'Place of Usual Residence', 'Births','Estimated Resident Population'
birth_pop_2012_df = birth_data_2012_df.loc[:,["Place of Usual Residence","Births", "Estimated resident population"]]
#birth_pop_2012_df
```

```python
#Remove the commas from 'Births' and 'Estimated Resident Population'  
birth_pop_2019_df['Births'] = birth_pop_2019_df['Births'].str.replace(',', '')
birth_pop_2012_df['Births'] = birth_pop_2012_df['Births'].str.replace(',', '')
birth_pop_2019_df['Estimated resident population'] = birth_pop_2019_df['Estimated resident population'].str.replace(',', '')
birth_pop_2012_df['Estimated resident population'] = birth_pop_2012_df['Estimated resident population'].str.replace(',', '')
```

```python
#Check the data types of the columns
birth_pop_2019_df.dtypes
```

```python
#Check that the commas have been removed
birth_pop_2012_df
```

```python
#In order to calculate the crude birth rate, change the data types of the values in 'Births' and 'Estimated Resident Population' from object to integer 
birth_pop_2019_df['Births'] = birth_pop_2019_df['Births'].astype(int)
birth_pop_2012_df['Births'] = birth_pop_2012_df['Births'].astype(int)
birth_pop_2019_df['Estimated resident population'] = birth_pop_2019_df['Estimated resident population'].astype(int)
birth_pop_2012_df['Estimated resident population'] = birth_pop_2012_df['Estimated resident population'].astype(int)
```

```python
#Re-check the data types of the columns
birth_pop_2012_df.dtypes
```

```python
#Calculate the Crude Birth Rate for 2019
birth_pop_2019_df["Crude Birth Rate"] = birth_pop_2019_df["Births"]/birth_pop_2019_df["Estimated resident population"]*1000
#birth_pop_2019_df

#Calculate the Crude Birth Rate for 2012
birth_pop_2012_df["Crude Birth Rate"] = birth_pop_2012_df["Births"]/birth_pop_2012_df["Estimated resident population"]*1000
birth_pop_2012_df.head()
```

```python
#Remove 'Australia' total from the dataframe
clean_birth_pop_2019_df = birth_pop_2019_df.loc[birth_pop_2019_df["Place of Usual Residence"]!="AUSTRALIA",:]
```

```python
#Check that 'Australia' has been removed
clean_birth_pop_2019_df.head()
```

```python
#Remove 'Australia' total from the dataframe
clean_birth_pop_2012_df = birth_pop_2012_df.loc[birth_pop_2012_df["Place of Usual Residence"]!="AUSTRALIA",:]
```

```python
#Plot crude birth rate for 2012
#crude_birth_rate_bar_2012 = clean_birth_pop_2012_df.plot.bar(x='Place of Usual Residence', y='Crude Birth Rate',title="Births in 2012 per 1000 people", rot=90)
#crude_birth_rate_bar_2012
```

```python
#Merge the 2012 and 2019 crude birth rate dataframes
crude_birth_rate_complete = pd.merge(clean_birth_pop_2012_df, clean_birth_pop_2019_df, how="left", on=["Place of Usual Residence"])
```

```python
#Get needed data for analysis
crude_birth_rate_2012_2019 = crude_birth_rate_complete.loc[:,["Place of Usual Residence","Crude Birth Rate_x", "Crude Birth Rate_y"]]
#crude_birth_rate_2012_2019

#Rename columns in the dataframe
crude_birth_rate = crude_birth_rate_2012_2019.rename(columns={"Crude Birth Rate_x": "2012", "Crude Birth Rate_y": "2019"})
crude_birth_rate.head()
```

```python
#Plot crude birth rate for 2012 and 2019 for each State capital and regional area and save figure
crude_birth_rate.plot.bar(x='Place of Usual Residence', title="Births (per 1000 people)",rot=90,).get_figure().savefig('Crude Birth Rate.png')
```

```python
#Calculate the difference between crude birth rate for 2019 and 2012
crude_birth_rate["Difference"] = crude_birth_rate["2019"]-crude_birth_rate["2012"]
#crude_birth_rate

#Plot the difference between crude birth rate for 2019 and 2012 and save the figure
crude_birth_rate.plot.bar(x='Place of Usual Residence', y='Difference', title="Differences in Crude Birth Rate between 2012 and 2019",legend=False,rot=90).get_figure().savefig('Differences in Crude Birth Rate.png')
```
