```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
%matplotlib inline
```

```python
# Read in csv
birth_path = 'Database 1.xlsx - Sheet2_starter.csv'
birth_data_df = pd.read_csv(birth_path)

birth_data_df
```

```python
birth_data_2015_df = birth_data_df.loc[birth_data_df['Year'] == 2015]
birth_data_2015_df
```

```python
birth_data_syd_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Sydney"]
birth_data_syd_df
```

```python
birth_data_nsw_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of NSW"]
birth_data_nsw_df
```

```python
birth_data_aus_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "AUSTRALIA"]
birth_data_aus_df
```

```python
birth_data_greater_brisbane_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Greater Brisbane"]

birth_data_greater_brisbane_df
```

```python
birth_data_rest_of_qld_df = birth_data_df.loc[birth_data_df['Place of Usual Residence'] == "Rest of Qld"]

birth_data_rest_of_qld_df
```

```python
number_column = birth_data_greater_brisbane_df.loc[:,'Median age of mother']
numbers_greater_brisbane = number_column.values


print(numbers_greater_brisbane)
```

```python
number_column_rest_of_qld = birth_data_rest_of_qld_df.loc[:,'Median age of mother']
numbers_rest_of_qld = number_column_rest_of_qld.values


print(numbers_rest_of_qld)
```

```python
number_column_year = birth_data_greater_brisbane_df.loc[:,'Year']
year = number_column_year.values
print(year)
```

```python

#Determine the median age of mother over the period

# Plot both on the same chart

plt.plot(year, numbers_greater_brisbane)
plt.plot(year, numbers_rest_of_qld)
plt.show()

```

```python
#Set legend
GreaterBrisbane, = plt.plot(year, numbers_greater_brisbane, color="blue", linewidth=1, label="Greater Brisbane")
RestofQld, = plt.plot(year, numbers_rest_of_qld, color="Red", linewidth=1, label="Rest of Qld")
plt.legend(handles=[GreaterBrisbane, RestofQld], loc="best")


# Give our graph axis labels

plt.xlabel("YEAR")
plt.ylabel("Median age of mother")
```
