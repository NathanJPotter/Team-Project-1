```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#import scipy.stats as st
#from scipy.stats import linregress
#from sklearn import linear_model
#import statsmodels.api as sm
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
#  data files
house_price_city = "Resources/house_price_city.csv"
births = "Resources/births_number.csv"
```

```python
# Read the data
house_price =  pd.read_csv(house_price_city)
births = pd.read_csv(births)
# house_price["Births"] = np.nan
data_df = pd.merge(house_price, births, on=['Year','Location'])
```

```python
data_df
```

```python
colors = {"Urban":"orchid","Rural":"lightgreen"}
plt.style.use("seaborn")

scatter_plot_df = data_df[["Year","Price","Type"]]

urban_set = data_df[data_df.Type == "Urban"]
rural_set = data_df[data_df.Type == "Rural"]

ax = urban_set.plot.scatter(x="Year",y="Price",
                       edgecolor='dimgrey',
                       s=(urban_set.Births/250),
                       color=colors["Urban"],
                       label="Urban",
                       linewidth = 1,)


rural_set.plot.scatter(x="Year",y="Price",
                       edgecolor='dimgrey',
                       s=(rural_set.Births/250),
                       color=colors["Rural"],
                       label="Rural",
                       linewidth = 1,
                       ax=ax)

plt.yticks(np.arange(200000, 1000000, 50000))
plt.legend(title="Types")
plt.title("House Prices and Birth Rate - Urban vs Rural")

plt.show()
```
