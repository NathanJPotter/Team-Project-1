```python
birth_data_2019_df = birth_data_df.loc[birth_data_df['Year'] == 2019]
#birth_data_2019_df

birth_data_2012_df = birth_data_df.loc[birth_data_df['Year'] == 2012]
#birth_data_2012_df
```

```python
birth_pop_2019_df = birth_data_2019_df.loc[:,["Place of Usual Residence","Births", "Estimated resident population"]]
#birth_pop_2019_df

birth_pop_2012_df = birth_data_2012_df.loc[:,["Place of Usual Residence","Births", "Estimated resident population"]]
birth_pop_2012_df
```

```python
#birth_pop_2019_df["Births"] = birth_pop_2019_df["Births"].astype(str).astype(int)
#birth_pop_2019_df["Estimated resident population"] = pd.to_numeric(birth_pop_2019_df["Estimated resident population"])
birth_pop_2019_df['Births'] = birth_pop_2019_df['Births'].str.replace(',', '')
birth_pop_2012_df['Births'] = birth_pop_2012_df['Births'].str.replace(',', '')

birth_pop_2019_df['Estimated resident population'] = birth_pop_2019_df['Estimated resident population'].str.replace(',', '')
birth_pop_2012_df['Estimated resident population'] = birth_pop_2012_df['Estimated resident population'].str.replace(',', '')
#birth_pop_2019_df.dtypes
#births_integer = birth_pop_2019_df["Births"].str.replace(',', '').astype(float).astype(int)
```

```python
birth_pop_2019_df['Births'] = birth_pop_2019_df['Births'].astype(int)
birth_pop_2012_df['Births'] = birth_pop_2012_df['Births'].astype(int)

birth_pop_2019_df['Estimated resident population'] = birth_pop_2019_df['Estimated resident population'].astype(int)
birth_pop_2012_df['Estimated resident population'] = birth_pop_2012_df['Estimated resident population'].astype(int)
```
